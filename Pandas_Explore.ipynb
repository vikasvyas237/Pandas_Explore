## <font color = 'black'>**Basic Analysis** </font>
**value_counts()** returns the series object counting all the unique values. <br> <br>
**NOTE:** The first value is the most recurring event and it goes on with other value sequence. This can be reversed by setting **ascending = True** <br>
**dropna = True** by default and we will not get the count of the N/A but if the dataframe has more N/A values this can be misleading and we need to turn that to **False**

```python
print(oo.Edition.value_counts(), '\n\n') #this will tell that the year in which there were most medals 
#presented to the least medals presented

#this is print the number of medals won by genders and as ascending = True 
#this will print the results in ascening sequence
print(oo.Gender.value_counts(ascending=True, dropna = False))
```

**sort_values()** sorts the values in a series. <br>
We can sort multiple series in a particular order by DataFrame.sort_values[['Series1', 'Series2']]


```python
#this will print the names of all the athletes by their name and be assigned to a new series "ath"
ath = oo.Athlete.sort_values()
print(ath)
```

```python
#this will sort the values of the dataframe firstly by the Edition
#so the Edition will begin from 1896
#after that it will sort the athlete column in alphabetical order
oo.sort_values(by=['Edition','Athlete'])
```

## <font color = 'black'>**Basic Plotting** </font>

```python
import matplotlib.pyplot as plt #standard way to include the matplotlib.pyplot library
%matplotlib inline 
#this important function will display the plots below the code and 
#also save them in the document
```

**plot()** is used to display the graph. By default it is a **line** graph. If we want to specifiy the graph then we can enter <br> <br>
<li>plot(kind='bar')</li>
<li>plot(kind='pie')</li>
<li>plot(kind='barh') #horizontal bar graph</li> 

**NOTE:** It is important to know when to use which type of graph on the basis of the data and the problem we are trying to solve <br> <br>
**Line Plot** is best for tracking changes over a period of time <br>
**Bar Plot** is used to compare between a different groups <br>
**Pie Plot** is used to compare the parts of the whole and if they are not changing over time or if we are comparing them for a particular moment of time 

### <font color = 'black'>What were the different sports in the first olympics. We need to plot them with the most suitable graph </font>

```python
first_oly = oo[oo['Edition'] == 1896]
first_oly.head()
```

```python
#NOTE: The semicolon is always used at the end so that the header code is not printed
#in this instance - <AxesSubplot:>
#figsize is passed as a tuple -> plot(figsize(width,height)) in inches
first_oly['Sport'].value_counts(ascending=True).plot(kind='bar',color='darkblue',figsize=(8,4));
```

## <font color ='black'>**Colormaps** </font> 
The idea behind choosing a good colormap is to represent the data in a good manner <br>
**Types of Colormaps** <br>
<li>Sequential - It is used where there is ordering (change in lightness over a single hue)</li>
<li>Diverging - Must be used when the plotted info diverges around a middle value</li><li>Qualitative - Used where there isn't any ordering (different colors used for types)</li>

```python
first_oly['Sport'].value_counts().plot(kind='pie', colormap='Pastel2');
```

## <font color='black'>**Seaborn Basic Plotting**</font>
It is a visualizaiton library based on Matplotlib. It is used to make attractive plots. Seaborn is a complement to Matplotlib. It integrates well with pandas. It is used for more advanced plots or where we need to plot categorical r statistical data. 

```python
import seaborn as sns #this is the standard way of importing 
```

**How many medals have been won by men and women in the history of Olympics. How many gold, silver & bronze medals were won according to each gender** <br><br>
**countplot()** is an important method of the seaborn library <br>
**countplot(x**= 'data on the x-axis', **data**='dataframe or the source of the data', **hue**='categorical variables', **orde**r='sequence of the categorical variables', **palette** = 'different levels of the categorical variables')

```python
sns.countplot(x='Medal', data=oo, hue='Gender');
```
